```python
#08 Leap Year Checker
```

```python
year = int(input("please enter a year:- "))

if year % 400 == 0:
  print("its a leap year")
elif year % 4 == 0 and year % 100 != 0:
  print("its a leap year")
else:
  print("its not a leap year")
```
